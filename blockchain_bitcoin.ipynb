```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
wallet_data=pd.read_csv('wallet.csv')
wallet=wallet_data['my-wallet-n-users']
mempool_size_data=pd.read_csv('mempool-size.csv')
mempool_size=mempool_size_data['mempool-size']
difficulty_data=pd.read_csv('difficulty.csv')
difficulty=difficulty_data['difficulty']
cost_per_transaction_data=pd.read_csv('cost-per-transaction.csv')
cost_per_transaction=cost_per_transaction_data['cost-per-transaction']
price_data=pd.read_csv('market-price.csv')
price=price_data['market-price']
hashrate_data=pd.read_csv('hash-rate.csv')
hashrate=hashrate_data['hash-rate']
miners_revenue_data=pd.read_csv('miners-revenue.csv')
miners_revenue=miners_revenue_data['miners-revenue']
total_transaction_data=pd.read_csv('transactions-total.csv')
total_transaction=total_transaction_data['n-transactions-total']
utxo_data=pd.read_csv('utxo-count.csv')
utxo=utxo_data['utxo-count']
unique_address_data=pd.read_csv('unique-addresses.csv')
unique_address=unique_address_data['n-unique-addresses']
transaction_fees_data=pd.read_csv('transaction-fees.csv')
transaction_fees=transaction_fees_data['transaction-fees']
fees_usd_per_transaction_data=pd.read_csv('fees-usd-per-transaction.csv')
fees_usd_per_transaction=fees_usd_per_transaction_data['fees-usd-per-transaction']
total_bitcoin_data=pd.read_csv('total-bitcoins.csv')
total_bitcoin=total_bitcoin_data['total-bitcoins']
avg_block_size_data=pd.read_csv('avg-block-size.csv')
avg_block_size=avg_block_size_data['avg-block-size']
confirmation_time_data=pd.read_csv('median-confirmation-time.csv')
confirmation_time=confirmation_time_data['median-confirmation-time']
tx_per_day_data=pd.read_csv('n-transactions.csv')
tx_per_day=tx_per_day_data['n-transactions']
tx_per_block_data=pd.read_csv('n-transactions-per-block.csv')
tx_per_block=tx_per_block_data['n-transactions-per-block']
```

```python
X=pd.concat([wallet,difficulty,cost_per_transaction,hashrate,miners_revenue,
             total_transaction,utxo,unique_address,
             transaction_fees,fees_usd_per_transaction,total_bitcoin],axis=1)
Y=price
```
